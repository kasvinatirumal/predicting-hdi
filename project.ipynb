### Data Preprocessing

We first remove the last two rows from our main dataset, which do not contain relevant data.

```python
econ_indicators = econ_indicators.iloc[:-2]
```

The dataset is then cleaned, merged, and pivoted, to ensure it is in the aforementioned desired format.

```python
def clean_merge_and_pivot(df):    
    """
    Takes in a dataset df and filters it to only include rows
    with the following nine economic indicators found under the
    "WEO Subject Code" column: "NGDPDPC", "NID_NGDP", "NGSD_NGDP",
    "PCPIPCH", "TM_RPCH", "TX_RPCH", "LUR", "LP", and "PPPEX".
    The dataset is then transformed into a long format with "Year"
    as a single column, and then pivoted back to a wide format where
    each WEO Subject Code and their respective values become separate
    columns. The resulting dataset with its index reset is then returned. 
    Assumes that the given df includes the columns "WEO Subject Code",
    "ISO", "Country", and multiple year columns.
    """
    codes = ["NGDPDPC", "NID_NGDP", "NGSD_NGDP", "PCPIPCH", "TM_RPCH",
             "TX_RPCH", "LUR", "LP", "PPPEX"] 
    # filters the dataset to only include rows with the WEO Subject Codes
    # listed in codes
    df = df.loc[df["WEO Subject Code"].isin(codes), :]

    # sets up relevant years as string for analysis
    years = [str(year) for year in range(2010,2023)]
    years_str = [year for year in years if year in df.columns]
    
    # converts the dataset from wide to long format
    df_melted = df.melt(
        id_vars=["ISO","Country","WEO Subject Code"],
        value_vars=years_str,
        var_name="Year", value_name="Value"
    )

    # converts the dataset from long to wide format
    df_pivoted = df_melted.pivot(
        index=["ISO","Country","Year"], 
        columns="WEO Subject Code", values="Value"
    )
    return df_pivoted.reset_index() # resets index 


econ_df = clean_merge_and_pivot(econ_indicators)
```

Next, we remove non-numeric characters from columns representing numeric values, and convert them into a numeric data type.

```python
def convert_numeric(df):
    """
    Takes in a dataset df, and processes all columns starting from
    the third columns onwards. Non-numeric characters from values
    within these columns are removed, and converted into a numeric
    data type. Values that cannot be converted to a numeric data 
    type are set to NaN. The resulting dataset is returned. 
    Assumes that the given df has at least 3 columns.
    """
    for col in df.columns[2:]: 
        # strips non-numeric characters from the column values
        df[col] = df[col].replace("[^0-9.]", "", regex=True)
        df[col] = pd.to_numeric(df[col], errors="coerce")
    return df


econ_df_num = convert_numeric(econ_df)
```

We then filter our dataset so that it only contains years between 2010 and 2022 (inclusive), which is our aforementioned range of analysis.

```python
def filter_relevant_years(df):
    """
    Takes in a dataset df, and filters the dataset to include only
    rows where values under the "Year" column fall within the range
    2010 to 2022 (inclusive). The filtered dataset is returned.
    Assumes that the given df contains a column named "Year".
    """
    years_int = [year for year in range(2010,2023)] 
    return df.loc[df["Year"].isin(years_int)]


co2_filtered = filter_relevant_years(co2_emissions)
hdi_filtered = filter_relevant_years(hdi)
unemp_filtered = filter_relevant_years(unemployment_rate)
```

Next, we clean the remaining 4 datasets by renaming their respective statistic columns to shorter names, which improves readability and simplifies future analysis. 

```python
def rename_col(df, old_name, new_name):
    """
    Takes in a dataset df, old_name, and new_name, and renames
    a column in the dataset from the given old_name to the given
    new_name. The renamed dataset is returned.
    Assumes the given dataset has a column with the given old_name.
    """
    return df.rename(columns={old_name: new_name})


co2_renamed = rename_col(co2_filtered, "Annual CO₂ emissions (per capita)", "CO2")
hdi_renamed = rename_col(hdi_filtered, "Human Development Index", "HDI")
unemp_renamed = rename_col(unemp_filtered, "Unemployment, total " 
                "(% of total labor force) (modeled ILO estimate)", "UR")
cont_renamed = rename_col(continents, "World regions according to OWID", "Continent")
```

Finally, the main dataset is merged with the other 4 datasets using a left join. The `Per Capita CO2 Emissions`, `Human Development Index`, and `Unemployment Rate` datasets is joined onto the main dataset using both `Country` and `Year` as keys, while the `Continents` dataset will be merged with the main dataset, using only `Country` as the key.

Additionally, we replace the `LUR` column within the combined dataset with unemployment rate data from the `Unemplyment Rate` dataset, because of the large number of missing values within the `LUR` column of the main dataset.

```python
def combine_datasets(df1, df2, l_keys, r_keys=None):
    """
    Takes in two dataset, df1 and df2, and left join keys l_keys,
    and optional right join keys r_keys. The datasets df1 and
    df2 are merged using a left join on the given l_keys and
    r_keys. If r_keys is not provided, we assume that the left
    and right join keys are identical. The resulting merged dataset
    with the "Entity" column dropped is returned.
    Assumes that the df1 and df2 contains the columns listed in
    l_keys and r_keys respecitevely. If r_keys was not provided,
    assumes that df2 contains the columns listed in l_keys.
    """
    if r_keys is None:
        r_keys = l_keys
    return pd.merge(df1, df2, how="left", left_on=l_keys, right_on=r_keys).drop(columns="Entity")
    

combined_1 = combine_datasets(econ_df_num, co2_renamed, ["Country","Year"], ["Entity","Year"])
combined_2 = combine_datasets(combined_1, hdi_renamed, ["Country","Year"], ["Entity","Year"])
combined_3 = combine_datasets(combined_2, unemp_renamed, ["Country","Year"], 
                              ["Entity","Year"]).drop(columns="LUR")
combined_4 = combine_datasets(combined_3, cont_renamed, "Country", "Entity")
```

### RQ1: Changes in per capita GDP, CO2 emissions, and HDI of each country from 2010 to 2022

In this section, we examine the changes in per capita GDP, CO2 emissions, and HDI across countries worldwide from 2010 to 2022 (inclusive). The changes are visualized using a choropleth map for each metric over the given period.

```python
def plot_interactive_viz(df, col, title, file_name, legend=None):
    """
    Takes in a dataset df, column col, title, file_name and an optional 
    legend, creates an interactive choropleth map of the values from
    the given col for each country, and saves the plot as an html file.
    The map includes an animation of the data from the year 2010 to 2022
    (inclusive). A diverging color scale of BrBG is used, and the map
    is titled using the given title.
    If a legend is provided, it is used to label the color scale. If not,
    the color scale will share the same label as the given col.
    The map is titled "Worldwide Per Capita GDP (2010-2022)" (if Per
    Capita GDP is the given col).
    Assumes that the given dataset has the columns "Country", "Year",
    and the given col.
    Assumes the given file_name contains no spaces.
    """
    fig = px.choropleth(
        df,
        locations="Country",
        locationmode="country names",
        color=col,
        hover_name="Country",
        animation_frame="Year",
        color_continuous_scale=px.colors.diverging.BrBG,
        title=f"Worldwide {title} (2010-2022)"
    )
    if legend is not None:
        fig.update_layout(coloraxis_colorbar_title=legend)
    fig.write_html(f"{file_name}.html")
    

plot_interactive_viz(econ_df, "NGDPDPC", "Per Capita GDP", "per_capita_gdp", "GDP ($)")
plot_interactive_viz(econ_df, "CO2", "Per Capita CO2 Emissions", "per_capita_co2", "CO2 (tonnes)")
plot_interactive_viz(econ_df, "HDI", "Human Development Index", "hdi")
```

```python
def plot_line_chart(df):
    """
    Takes in a dataset df, and plots three line charts displaying
    the median values of Per Capita GDP, CO2 emissions, and HDI by
    continent from 2010 to 2022 (inclusive). Each plot displays
    six separate lines, each representing a different continent.
    The charts are arranged on a 2x2 grid, with two plots on the
    top row, and one on the left side of the bottom row.
    The titles of the plot from top to bottom, left to right, include
    "Per Capita GDP($) by Continent", "Per Capita CO2 Emissions (tonnes)
    by Continent", and "HDI by Continent". All three plots have the
    same x-axis label named "Year", while the y-axes are labeled 
    "Median Per Capita GDP ($)", "Median Per Capita CO2 Emissions
    (tonnes)" and "Median HDI" respectively.
    The resulting 2x2 figure is displayed. 
    Assumes the given df contains the columns "NGDPDPC", "CO2", "HDI",
    "Year" and "Continent".
    """
    cols = ["NGDPDPC", "CO2", "HDI"]
    titles = ["Per Capita GDP ($)", "Per Capita CO2 Emissions (tonnes)", "HDI"]
    data = df.groupby(["Continent", "Year"])[cols].median().reset_index()

    fig, axes = plt.subplots(2, 2, figsize=(13.5, 8))
    # flatten array to iterate over it
    axes = axes.flatten()
    for ax, col, title in zip(axes[:3], cols, titles):
        sns.lineplot(data=data, x="Year", y=col, ax=ax, hue="Continent")
        ax.set(
            title=f'{title} by Continent',
            xlabel="Year", ylabel=f"Median {title}"
        )
    # remove empty subplot
    fig.delaxes(axes[3])
    fig.suptitle("Line Plot of Each Metric by Continent", fontsize=14)
    plt.tight_layout()
    plt.show()


plot_line_chart(econ_df)
```

### RQ2: Statistical Analysis of Variation in HDI Across CO2 Emission Levels

In this section, we will perform a hypothesis test to determine whether there is a statistically significant difference in HDI values across different per capita CO2 emission levels in 2022. The hypotheses are as follows:
- Null hypothesis: HDI values do not differ across CO2 emission levels in 2022
- Alternative hypothesis: HDI values differ across CO2 emission levels in 2022

We start by imputing missing values within our dataset `econ_df`, using the median values of their respective continent for the given year. Instead of dropping these rows, we opted to impute missing values as we wanted to perserve valuable information from other columns, which would be especially imoprtant when training our machine learning models. 

```python
def impute(df):
    """
    Takes in a dataset df, and imputes missing values within numerical
    columns of the dataset by replacing them with the median value of
    the continent that this country belongs to for that given year. 
    The resulting dataset is returned. 
    Assumes the given df contains the columns "Continent" and "Year".
    """
    numerical_cols = df.select_dtypes(include=[np.number])
    for col in numerical_cols:
        group = df.groupby(["Continent", "Year"])[col]
        df[col] = group.transform(lambda x: x.fillna(x.median()))
    return df


# drops rows where values are still missing after imputation
econ_imputed = impute(econ_df).dropna() 
econ_df_2022 = econ_imputed[econ_imputed["Year"]==2022].copy()
```

Next, a new column named `CO2_Level` is added to the `econ_df_2022` dataset, that categorizes per capita CO2 emissions into Low, Medium, or High using its 33rd and 66th percentile thresholds.

```python
def categorize(df, col):
    """
    Takes in a dataset df and a column col, and creates a new
    column named "CO2_Level" (if the given col is CO2), then 
    categorizes values in the given col into three levels:
    Low, Medium or High. Values less than or equal to the 33th
    percentile are categorized as Low, values between the 33th
    and 66th percentile (inclusive) are categorized as Medium,
    and the rest are categorized as High.
    The resulting dataset with the new column is returned. 
    Assumes the given df contains the given col.
    """
    low = df[col].quantile(0.33)
    high = df[col].quantile(0.66)

    new_col = f"{col}_Level"
    df[new_col] = 'Low' 
    df.loc[(df[col] > low) & (df[col] <= high), new_col] = 'Medium'
    df.loc[df[col] > high, new_col] = 'High'
    return df


econ_2022 = categorize(econ_df_2022, "CO2")
```

A boxplot is then created to visualize the distribution of HDI values across each CO2 emission level category.

```python
def plot_hdi_boxplot(df, col, title):
    """
    Takes in a dataset df, a column col, and a title, and plots
    a boxplot displaying the distribution of HDI values for
    different categories of the given col, with the given title.
    The x-axis is labeled "CO2_Level" (if the given col is 
    CO2_level), and the y-axis is labeled "HDI".
    The categories on the x-axis are ordered as 'Low', 'Medium',
    and 'High'. The resulting boxplot is displayed.
    Assumes the given df contains the column "HDI" and the given col.
    """
    plt.figure(figsize=(6, 4.5))
    sns.boxplot(x=col, y="HDI", data=df, order=["Low", "Medium", "High"])
    plt.title(title)


plot_hdi_boxplot(econ_2022, "CO2_Level", "Boxplot of HDI against CO2 Emissions Level")
```

The box plot above again **confirms the positive correlation between per capita CO2 emissions and HDI**, where median HDI increases as CO2 emissions level increase from Low to High. Furthermore, the spread of HDI values appear to shrink as CO2 emission levels increase, as evidenced by the shorter whiskers. In addition, the Medium CO2_Level has the most outliers of three, while the Low CO2_Level has one, followed by High which has none. 

Overall, this visualization **provides us the intuition that HDI values may vary across CO2 emission levels**. To formally test this observation, we perform a one-way ANOVA test as shown below. 

```python
def one_way_anova(df, col):
    """
    Takes in a dataset df, and column col, and performs a one-way
    ANOVA test to compare the means of HDI values across different
    categories of the given column. The results of the ANOVA test
    is returned as a tuple that includes the F-statistic and p-value.
    Assumes the given df contains the column "HDI" and the given col.
    """
    samples = []
    for level, agg_values in df.groupby(col):
        hdi_values = agg_values["HDI"].values
        samples.append(hdi_values)
    return stats.f_oneway(*samples)


F_stat, p_val = one_way_anova(econ_2022, "CO2_Level")
print(f"F-statistic: {F_stat}")
print(f"P-value: {p_val}")
```

The ANOVA test results in a p-value of $8.38 \times 10^{-48}$, which is way smaller than the significance level of 0.05. Consequently, **we reject our null hypothesis** that HDI values do not differ across CO2 emission levels in 2022. Furthermore, our F-statistic is exceptionally large, reinforcing that HDI does indeed vary significantly across different per capita CO2 emission levels. This is likely because countries with higher per capita CO2 emissions tend to be more economically developed, and thus have invested more into human development, and therefore have higher HDI. 

This significant varation of HDI across different CO2 emission levels further suggests that **per capita CO2 emissions will be a key predictor of HDI**.

### RQ3: Predicting HDI using Economic Indicators

In this section, we aim to predict HDI using only economic indictors. Consequently, our feature matrix `X` will include the nine economic indicators and `Year`, while the target variable `y` consist of HDI values. `Year` was included as a predictor to allow the model to capture temporal changes in economic indicators and HDI throughout the time period. A train-test split of 60-40 is employed, where 60% of the data is used to train the regression models, and the remaining 40% for testing. 

```python
X = pd.get_dummies(econ_imputed.drop(["ISO", "Country", "Continent", "CO2", "HDI"], axis=1))
y = econ_imputed["HDI"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=2025)

lreg_model = LinearRegression()
xgb_model = xgb.XGBRegressor(random_state=25)
```

We start by training a basic linear regression model and an XGBoost model on the training set (`X_train`, `y_train`), and evaluate their performance on the test set `y_test` using the root mean squared error (RMSE).

```python
def basic_reg(X, y, model, name):
    """
    Fits a regression model of the given model type to the 
    given feature matrix X and given target vector y. The
    RMSE of the model's predictions on the test set is then
    calculated. The RMSE is then printed using the given name,
    and the trained model is returned. 
    Assumes y_test is defined prior to running this function.
    """
    reg = model.fit(X, y)
    print(f"{name} Error:", root_mean_squared_error(y_test, reg.predict(X_test)))
    return reg


lin_reg = basic_reg(X_train, y_train, lreg_model, "Lin Reg")
xgb_reg = basic_reg(X_train, y_train, xgb_model, "XGBoost")
```

The **basic XGBoost model performs better than the linear regression model**, as it achieves a smaller RMSE of 0.027, compared to 0.106 of the linear regression model. 

This difference in performance could be due to several factors, the first being that the relationship between HDI and its predictors may be non-linear. Since XGBoost is a tree-based model, it can capture non-linear relationships more effectively, whereas linear regression stricty assumes a linear relationship, even when interaction terms are added to the model. Furthermore, XGBoost is more robust to outliers, allowing it to differentiate between meaningful patterns and noise more effectively than linear regression. 

Next, we train a linear regression and XGBoost model using **recursive feature elimination (RFE)** to determine whether we can maintain the models' accuracy, while using fewer predictors. We want to train our models using the fewest possible predictors, so that users only need to provide minimal input to predict HDI. 

**We expect the RMSE for both models to increase**, as the models now have fewer predictors to learn from. 

```python
def rfe(X, y, model, name):
    """
    Performs recursive feature elimination (RFE) using a regression
    model of the given model type to select the top 5 features from
    the given feature matrix X and target vector y.
    The list of top 5 features and the RMSE of the model’s 
    predictions on the test set are then printed out using the
    given name on two separate lines, followed by a blank line.
    The list of top 5 features is returned.
    Assumes y_test is defined prior to running this function.
    """
    rfe = RFE(model, step=1, n_features_to_select=5, verbose=1)
    rfe.fit(X, y)
    rfe_features = X.columns[[r == 1 for r in rfe.ranking_]]
    
    print(f"{name} Features:", list(rfe_features))
    print(f"{name} RFE Error:", root_mean_squared_error(y_test, rfe.predict(X_test)))
    print()
    return list(rfe_features)    


lr_features = rfe(X_train, y_train, lreg_model, "Lin Reg")
xgb_features = rfe(X_train, y_train, xgb_model, "XGBoost")
```

The RMSE of the linear regression model trained using RFE is 0.143, which is expectedly higher than its previous RMSE of 0.106 using the basic model. Surprisingly, **the new XGBoost model achieves a smaller RMSE of 0.024 comapred to its previous value of 0.027**. This suggests that the XGBoost model performed better with fewer predictor (5 predictors), than when it had access to all predictors within `X_train`. One possible explanation is that when XGBoost had access to more predictors, it may have overfitted the data. Reducing the number of predictors allowed the model to learn more important patterns, which led to increased performance and predictive accuracy. 

Another difference is that the linear model selected `Year`, `NGSD_NGDP`, `NID_NGDP`, `TM_RPCH` and `UR` as the 5 most important predictors, whereas the XGBoost model chose `LP`, `NGDPDPC`, `NID_NGDP`, `PPPEX` and `UR`. **Both models identified `NGDPDPC`, `NID_NGDP` and `UR` as top predictors, suggesting that these factors are most important economic indicators when prediciting a country's HDI.** 

Next, we want to improve the accuracy of our XGBoost model by **tuning its hyperparameters**. More specifically, the following hyperparamters will be tuned: *learning rate*, *maximum depth*, *minimum child weight*, and *column sample by tree*. This is achieved through cross-validation, as demonstrated below. Note that the XGBoost model will be trained solely on `xgb_features`, which were the top 5 predictors identified by XGBoost using RFE.

**We expect the RMSE of the hyperparameter-tuned model to be lower than the model trained using RFE.**

```python
def cross_validation_xgb(features, X, X_train, X_test, y):
    """
    Takes in a list of features, and subsets the given matrices
    X, X_train and X_test to only include the given features.
    Hyperparamter tuning is then performed using cross-validation
    on an XGBoost regression model, with the following parameters
    being tuned: learning rate, maximum depth, minimum child weight,
    and column sample by tree. The model is trained on the subsetted
    X-train and given y. The best model is chosen based on the
    highest negative root mean squared error (RMSE) score. The mean
    score of the best model, the model's hyperparamters, and the most
    important features ranked by feature importance scores in
    descending order are printed. The trained XGBoost model with the
    tuned hyperparamters is returned. 
    """
    X = X[features]
    X_train = X_train[features]
    X_test = X_test[features]
    param_grid={
        "learning_rate"    : [0.05, 0.10, 0.15, 0.2, 0.3],
        "max_depth"        : [3, 4, 5, 6, 7, 8, 10],
        "min_child_weight" : [1, 3, 6, 8, 10],
        "colsample_bytree" : [0.3, 0.5, 0.7, 0.8]
    }
    search = GridSearchCV(
        estimator=xgb.XGBRegressor(random_state=25),
        param_grid=param_grid,
        scoring="neg_root_mean_squared_error",
        n_jobs=-1, 
        verbose=0,
    )
    search.fit(X_train, y)
    print("Mean score for best model:", search.best_score_)
    reg = search.best_estimator_
    print("Best model:", reg)
    
    # Prints feature importance scores in descending order
    fi_scores = pd.Series(reg.feature_importances_, index=features).sort_values(ascending=False)
    print("Most Important Features:\n" + fi_scores.to_string())
    return reg


xgb_reg = cross_validation_xgb(xgb_features, X, X_train, X_test, y_train)
```

```python
print("XGBoost CV Error:", root_mean_squared_error(y_test, xgb_reg.predict(X_test[xgb_features])))
```

**The hyperparamter-tuned XGBoost model achieves an RMSE score of 0.026, which is surprisingly higher than the RMSE of 0.024 achieved through RFE**. The new RMSE is slightly lower than the basic XGBoost model which used all avaliable predictors, and achieved an RMSE of 0.027. 

**Therefore, the most accurate model for predicting HDI using economic indicators is the XGBoost model trained using RFE, `xgb_features`.**

However, it is also important to note that the difference between the RMSE of the XGBoost models trained using RFE and hyperparameter tuning via cross validation is minimal, at just 0.003. This suggests that **there is no significant difference between the two methods**, and depending on the random state of the XGBoost model or different train-test splits, the hyperparameter-tuned model could outperform RFE instead.

## Implications and Limitations

From this project, we learned that per capita GDP, CO2 emissions, and HDI are positively correlated, and that both economic indicators and CO2 emissions serve as significant predictors of HDI. 

The results of this project are valuable for those interested in improving a country's HDI, which includes academics, economists, policymakers, and local, state and governmental agencies. This is because stakeholders can use the predictive model to evaluate how changes in the values of different key predictors of HDI affect HDI, and select the optimal combination of predictor values that maximizes HDI while minimizing environmental impact and promoting economic growth. Policies can then be implemented to align real-word indicator values to match this optimal combination, which ultimately leads to a reliable improvement in HDI. 

However, it is important to consider these results alongside the limitations of this analysis, which includes:
1. **Possible misinterpretation of results**

    Our analysis revealed a moderate positive correlation between HDI and per capita CO2 emissions, suggesting that higher CO2 emissions are associated with higher HDI. While this may seem counterintuitive since we expect increased pollution to reduce quality of life, and thus lower HDI, this correlation actually reflects the indirect relationship between economic activity and CO2 emissions, not the intentional increase of emissions. Consequently, these findings could be misused by those seeking to harm the environment, potentially leading to policies that increase CO2 emissions under the false belief that they improve HDI.

2. **Reliability of data**

    The main dataset used within this analysis, `Economic Indicators`, was sourced from the IMF website. This presents a limitation because the IMF works closely with governments worldwide to foster global monetary corporation. This monetary relationship incentivizes countries to misreport their data in ways that align with their economic interests, especially since these numbers affect the financial assistance they receive from the IMF. As a result, the dataset may be inaccurate or biased, which may have skewed the strength of correlation observed between per capita GDP, CO2 emissions and HDI. Consequently, policies derived from this analysis may not be as effective as projected. For future analysis, we will cross-check the IMF data with data from other sources such as the World Bank, to ensure that they are consistent and accurate.

3. **Completeness of data**

    The `Economic Indicators` dataset contained a significant number of missing values, which were addressed through imputation, by using the median value of each continent for its respective year. However, it is important to consider that missing values were most prevalent among smaller and less developed countries. As a result, these countries should use the results of this analysis with caution, as the imputed data used to train the machine learning models may not best represent their economic and social trends. Conversely, this analysis is likely more accurate for developing and developed countries, since their data were typically more complete. For future analysis, instead of using the continent’s median for imputation, we will cross-check missing values with other databases to identify an exact value.

4. **Simplification of analysis**

    Our analysis was simplified by selecting only nine indicators out of the 44 available within the original `Economic Indicators` dataset. This selection was somewhat biased, as it was based on personal macroeconomic knowledge. Thus, it is likely that there may have been other important predictors of HDI that were unintentionally excluded from the analysis, which diminishes the accuracy of our findings. Thus, for future analysis, we will attempt to consider all 44 indicators, and identify the actual 5 most important predictors of HDI from them. This eliminates the possibility of missing important variables that correlate with HDI.  

5. **Absence of `Year` as a predictor in the final model**

    When looking at the top 5 predictors of HDI identified by the XGBoost model that included `CO2` as a predictor, it is important to note that `Year` was not selected as a top 5 predictor. Consequently, `Year` was excluded as a parameter for users to input when estimating HDI. However, given that we expect the values of economic indicators, CO2 emissions, and HDI to change on a year-to-year basis, it seems illogical to not include `Year` as a parameter. Hence, for future analysis, the importance of `Year` as a predictor will be evaluated, to better understand our model and strengthen the analysis within this project.
